```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('n_movies.csv')
df.head()
```

### Data Understanding

```python
df.info()
df.describe()
df.isnull().sum()
```

2. Handle Missing Values

```python
# Certificate 
df['certificate'] = df['certificate'].fillna('Unknown')

# Genre 
df['genre'] = df['genre'].fillna('Unknown')

# Duration 
df['duration'] = df['duration'].fillna(df['duration'].mode()[0])

# Rating 
df['rating'] = df['rating'].fillna(df['rating'].mean())

# Votes 
df['votes'] = df['votes'].str.replace(',', '')
df['votes'] = pd.to_numeric(df['votes'], errors='coerce')
df['votes'] = df['votes'].fillna(df['votes'].median())

# Year 
df = df.dropna(subset=['year'])
```

3. Correct Data Types

```python
# Year
df['year'] = df['year'].str.extract('(\d{4})')
df['year'] = pd.to_numeric(df['year'], errors='coerce')
# Duration
df['duration'] = df['duration'].str.replace(' min', '')
df['duration'] = pd.to_numeric(df['duration'], errors='coerce')

df.info()
```

4. Feature Engineering + Filtering

```python
# Creating New Columns

# Rating Category
def rating_category(x):
    if x >= 8:
        return 'Excellent'
    elif x >= 6:
        return 'Good'
    else:
        return 'Average'

df['rating_category'] = df['rating'].apply(rating_category)

# Extracting Main Genre
df['main_genre'] = df['genre'].str.split(',').str[0]

# Popularity Score (advanced feature)
df['popularity'] = df['rating'] * df['votes']
```

```python
# Applying Filters

# Movies after 2000
df = df[df['year'] >= 2000]

# Removing very low vote movies
df = df[df['votes'] > 1000]
```

```python
# Final Clean Check

df.info()
df.head()
```

## Data Visualisation


```python
# ── Global Styling ────────────────────────────────────────────────────────────
sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams.update({'figure.dpi': 120, 'font.size': 11})
```

### Plot 1 – Bar Chart: Top 10 Genres by Number of Titles

```python
genre_counts = (
    df['main_genre']
    .value_counts()
    .head(10)
    .reset_index()
)
genre_counts.columns = ['Genre', 'Count']

fig, ax = plt.subplots(figsize=(10, 5))

colors = sns.color_palette('Blues_d', 10)
ax.barh(genre_counts['Genre'][::-1], genre_counts['Count'][::-1], color=colors)

for i, (count, _) in enumerate(zip(genre_counts['Count'][::-1], genre_counts['Genre'][::-1])):
    ax.text(count + 5, i, str(count), va='center', fontsize=9)

ax.set_xlabel('Number of Titles')
ax.set_title('Top 10 Genres by Number of Titles on Netflix', fontsize=13, fontweight='bold')
ax.set_xlim(0, genre_counts['Count'].max() * 1.12)

plt.tight_layout()
plt.savefig('plot1_bar_genres.png', bbox_inches='tight')
plt.show()
```

### Plot 2 – Pie Chart: Rating Category Distribution

```python
cat_counts = df['rating_category'].value_counts()

fig, ax = plt.subplots(figsize=(7, 6))
wedges, texts, autotexts = ax.pie(
    cat_counts.values,
    labels=cat_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=sns.color_palette('Set2', len(cat_counts)),
    wedgeprops={'edgecolor': 'white', 'linewidth': 1.5}
)
for at in autotexts:
    at.set_fontsize(10)

ax.set_title('Netflix Content by Rating Category', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('plot2_pie_rating_category.png', bbox_inches='tight')
plt.show()
```

### Plot 3 – Histogram: Distribution of IMDb Ratings

```python
fig, ax = plt.subplots(figsize=(9, 5))
ax.hist(
    df['rating'],
    bins=30,
    color='steelblue',
    edgecolor='white',
    linewidth=0.6
)

ax.set_xlabel('IMDb Rating')
ax.set_ylabel('Number of Titles')
ax.set_title('Distribution of IMDb Ratings for Netflix Titles', fontsize=13, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))

excellent = df[df['rating_category'] == 'Excellent']
good      = df[df['rating_category'] == 'Good']
average   = df[df['rating_category'] == 'Average']

plt.scatter(excellent['rating'], excellent['votes'], label='Excellent', alpha=0.7)
plt.scatter(good['rating'], good['votes'], label='Good', alpha=0.7)
plt.scatter(average['rating'], average['votes'], label='Average', alpha=0.7)

plt.xlabel("Rating")
plt.ylabel("Votes")
plt.title("Rating vs Votes by Category", fontsize=13, fontweight='bold')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
year_counts = df['year'].value_counts().sort_index()

plt.figure(figsize=(10,5))

plt.plot(year_counts.index, year_counts.values, marker='o')

plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.title("Number of Movies Released Over Time", fontsize=13, fontweight='bold')

plt.tight_layout()
plt.show()
```
